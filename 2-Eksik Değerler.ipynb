```python
# Fonksiyona çevirdik.
def missing_values_table(df,na_name=False):
    na_columns = [col for col in df.columns if df[col].isnull().sum()>0] # Eksik değerleri seçtik
    
    n_miss = df[na_columns].isnull().sum().sort_values(ascending=False) # Sırala
    
    ratio = (df[na_columns].isnull().sum() / df.shape[0] * 100).sort_values(ascending=False) # Oran
    
    missing_df = pd.concat([n_miss, np.round(ratio,2)], axis=1, keys=["n_miss","ratio"]) # missing_df adında bir df'de birleştir
    
    print(missing_df,end="\n")
    
    if na_name: # İsimleri ver 
        return na_columns
```

```python
# cat_cols, cat_but_car
def grab_col_names(df,cat_th=10,car_th=20):
    cat_cols = [col for col in df.columns if df[col].dtypes == "O"]
    num_but_cat = [col for col in df.columns if df[col].nunique() < cat_th and df[col].dtypes != "O"]
    cat_but_car = [col for col in df.columns if df[col].nunique() > cat_th and df[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    
    #numumerik
    num_cols = [col for col in df.columns if df[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]
    
    print(f"Observations: {df.shape[0]}")
    print(f"Variables: {df.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}") # Kategorik
    print(f"num_cols: {len(num_cols)}") # Numerik
    print(f"cat_but_car: {len(cat_but_car)}") # Kategorik ama kardinal
    print(f"num_but_cat: {len(num_but_cat)}")
    return cat_cols, num_cols, cat_but_car

df = load()
```

```python
cat_cols, num_cols, cat_but_car = grab_col_names(df)
num_cols = [col for col in num_cols if col not in "PassengerId"]
```

```python
# Bu durumda 2 sınıfa sahip olan kategorik değişkenlerin 1.sınıfını atacak 2.sınıfını tutacak. Gözlemler 1 veya 0 değerini alır.
# 2 sınıf veya daha çok sınıfa sahip olan kategorik değişkenleri numerik bir şekilde ifade etmek.
dff = pd.get_dummies(df[cat_cols+num_cols],drop_first=True)
```

```python
# Değişkenlerin standartlaştırılması
# Değerleri 0 1 arasında değiştir
scaler = MinMaxScaler()
dff = pd.DataFrame(scaler.fit_transform(dff),columns=dff.columns)
dff.head()
```

```python
# knn'in uygulanması
# Bu bize makine öğrenmesi yöntemi ile tahmine dayalı bir şekilde eskik değerleri doldurma imkanı sağlar.
from sklearn.impute import KNNImputer
```

```python
# Uzaklık temelli bir yöntem olarak komşulara bakar.
# Komşulardan ilgili eksikliğe sahip gözlem birimine en yakın komşuların örneğin; 5 tanesini bulur.
# Bunların dolu olan gözlemlerinin ortalamasını buraya atar.
imputer = KNNImputer(n_neighbors=5)
dff = pd.DataFrame(imputer.fit_transform(dff),columns=dff.columns)
dff.head()
```

```python
dff = pd.DataFrame(scaler.inverse_transform(dff),columns=dff.columns)
dff.head()
```

```python
# Tahmine dayalı bulduğumuz değerleri df içinde yeni bir değer atadık.
df["age_imputed_knn"] = dff[["Age"]]

df.loc[df["Age"].isnull(), ["Age","age_imputed_knn"]]
```
